```python
import folium

def tampilkan_linestring_pada_peta(data_linestring):
    # Buat peta dengan lokasi awal berdasarkan rata-rata koordinat dari data linestring
    avg_lat = data_linestring.geometry.apply(lambda x: x.centroid.y).mean()
    avg_lon = data_linestring.geometry.apply(lambda x: x.centroid.x).mean()
    m = folium.Map(location=[avg_lat, avg_lon], zoom_start=15)

    # Tambahkan polyline untuk setiap linestring
    for linestring in data_linestring.geometry:
        folium.PolyLine(locations=[list(coord) for coord in linestring.coords], color='blue', weight=5, opacity=0.7).add_to(m)

    # Tambahkan GeoJson dari data GeoPandas
    folium.GeoJson(data_linestring.geometry.to_json(), name='Garis Jalan').add_to(m)

    return m
```
